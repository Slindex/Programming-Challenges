## Variables

Variables cannot be defined with starting numbers unless we use `_`.

Here I'm using `exec()` function because the variable definition is being executed at compilation, not at runtime. Order of code processing is:

`parsing -> compilation -> runtime`

`try/except` blocks only catches exceptions at runtime, so the `exec()` function allow us to define the variable at runtime in order to catch it.

```python
try:
    exec('7up = "Drink"') # Invalid
except:
    print("Invalid variable")

_7up = "Drink" # Valid
print(_7up)
```
